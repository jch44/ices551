```python
# Importer les bibliothèques nécessaires
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist
import scipy.cluster.hierarchy as sch

# Charger le dataset Iris
iris = datasets.load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names

# Créer un DataFrame pour les données
df = pd.DataFrame(X, columns=feature_names)
df['species'] = y

# Normaliser les données
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


# Appliquer la Classification Ascendante Hiérarchique
# Calculer la matrice des distances
dist_matrix = pdist(X_scaled, metric='euclidean')
print("dist_matrix:", dist_matrix.shape)
print(dist_matrix)

# Appliquer l'algorithme de liaison
linkage_matrix = linkage(dist_matrix, method='ward')
print("linkage_matrix:")
print(linkage_matrix)
# Tracer le dendrogramme
plt.figure(figsize=(20, 7))
dendrogram(linkage_matrix, labels=iris.target_names[y], leaf_rotation=90, leaf_font_size=10, show_contracted=True) #, truncate_mode="level", p=3)
plt.title('Dendrogramme de la Classification Ascendante Hiérarchique (CAH)')
plt.xlabel('Index des Échantillons')
plt.ylabel('Distance')
plt.show()
print(linkage_matrix.shape)
```
